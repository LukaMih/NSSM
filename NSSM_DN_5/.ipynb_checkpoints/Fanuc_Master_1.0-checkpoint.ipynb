```python
import cv2
import PIL.Image as Image
import matplotlib.pyplot as plt

import numpy as np
from PIL import Image

im = cv2.imread("./parts.jpg")
#plt.figure(figsize=(20,16))
#plt.imshow(im)
```

```python
im = cv2.cvtColor(im,cv2.COLOR_BGR2HSV)

lower_color = np.array([0,0,0])
upper_color = np.array([180,255,100])

im = cv2.inRange(im, lower_color, upper_color)
```

```python
params=cv2.SimpleBlobDetector_Params()

params.filterByArea=True
params.minArea = 1000
params.maxArea = 2000  

params.filterByCircularity = True
params.minCircularity = 0.7

params.filterByConvexity = False
params.filterByInertia = False
params.filterByColor = False

detector=cv2.SimpleBlobDetector_create(params)

keypoints = detector.detect(im) 


im2 = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
for i, kp in enumerate(keypoints):
    x, y = np.int32(kp.pt)
    
    #cv2.circle(im2, (x, y), radius=2, color=(255, 0, 0), thickness=3)
    
    cv2.putText(im2, str(i+1), (x + 5, y), cv2.FONT_HERSHEY_SIMPLEX, 
                1.5, (255, 0, 0), 3, cv2.LINE_AA)

# Prikaži rezultat
#plt.figure(figsize=(10, 6))
#plt.imshow(im2)
#plt.axis('off')

veliki = []

for keyPoint in keypoints:
    u=keyPoint.pt[1]
    v=keyPoint.pt[0]

    scale=1 
    x=(v)*scale       
    y=(u)*scale
    
    veliki.append([x,y])
```

```python
params=cv2.SimpleBlobDetector_Params()

params.filterByArea=True
params.minArea = 500
params.maxArea = 800  

params.filterByCircularity = True
params.minCircularity = 0.7

params.filterByConvexity = False
params.filterByInertia = False
params.filterByColor = False

detector=cv2.SimpleBlobDetector_create(params)

keypoints = detector.detect(im) 
    
im2 = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)

for i, kp in enumerate(keypoints):
    x, y = np.int32(kp.pt)
    
    #cv2.circle(im2, (x, y), radius=2, color=(255, 0, 0), thickness=3)
    
    cv2.putText(im2, str(i+1), (x + 5, y), cv2.FONT_HERSHEY_SIMPLEX, 
                1.5, (255, 0, 0), 3, cv2.LINE_AA)

#plt.figure(figsize=(10, 6))
#plt.imshow(im2)
#plt.axis('off')


mali = []

for keyPoint in keypoints:
    u=keyPoint.pt[1]
    v=keyPoint.pt[0]

    scale=1 
    x=(v)*scale       
    y=(u)*scale
    
    mali.append([x,y])
```

```python
print("Fanuc_Master je zaznal ",len(mali)," majhnih in ",len(veliki)," velikih krogov." )


for (x1, y1), (x2, y2) in zip(veliki, mali):
    cv2.line(im2, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 3)

plt.figure(figsize=(16,10))
plt.imshow(im2)
plt.axis('off')
plt.show()


koti = []
for (x1, y1), (x2, y2) in zip(veliki, mali):  
    dx = x2 - x1
    dy = y2 - y1

    kot_rad = np.arctan2(dy, dx)     
    kot_deg = np.degrees(kot_rad)    

    #print(kot_deg)
    koti.append(kot_rad)
    #print(kot_rad)
    
popravki = []
for kot in koti:
    popravek = kot -0.6071841235083298
    popravki.append(popravek)
    #print(popravek)
    
print("Robot dobi točke oz. središča prijema:")
n = 0
for (x, y) in veliki:
    n += 1
    print("\t\t tocka ", n, ":", x, y)
    
print("in popravke kotov:")
n = 0
for kot in popravki:
    n += 1
    print("\t\t tocka ", n, ":", kot*180/np.pi)
```
